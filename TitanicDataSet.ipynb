```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
sns.set(style="ticks")
```

```python
dataset = pd.read_csv('datasets/titanic.csv')
```

```python
print("Всего строк в датасете: ", dataset.shape[0])
print("Всего столбцов в датасете: ", dataset.shape[1])
dataset.dtypes
```

```python
fig, ax = plt.subplots(figsize=(10, 2))
sns.scatterplot(ax=ax, x="Age", y="Fare", data=dataset)
```

```python
fig, ax = plt.subplots(figsize=(10, 2))
sns.scatterplot(ax=ax, x="Age", y="Fare", data=dataset, hue="Survived")
```

```python
fig, ax = plt.subplots(figsize=(6,6)) 
sns.distplot(dataset['Age'])
```

```python
sns.jointplot(x="Age", y="Fare", data=dataset)
```

```python
sns.jointplot(x="Age", y="Fare", data=dataset, kind="hex")
```

```python
sns.jointplot(x="Age", y="Fare", data=dataset, kind="kde")
```

```python
sns.pairplot(dataset)
```

```python
dataset.corr()
```

```python
fig, ax = plt.subplots(1, 1, sharex='col', sharey='row', figsize=(10,5))
fig.suptitle('Корреляционная матрица')
sns.heatmap(dataset.corr(), ax=ax, annot=True, fmt='.1f')
```

```python
fig, ax = plt.subplots(1, 3, sharex='col', sharey='row', figsize=(15,5))
sns.heatmap(dataset.corr(method='pearson'), ax=ax[0], annot=True, fmt='.1f')
sns.heatmap(dataset.corr(method='kendall'), ax=ax[1], annot=True, fmt='.1f')
sns.heatmap(dataset.corr(method='spearman'), ax=ax[2], annot=True, fmt='.1f')
fig.suptitle('Корреляционные матрицы, построенные различными методами')
ax[0].title.set_text('Pearson')
ax[1].title.set_text('Kendall')
ax[2].title.set_text('Spearman')
```
